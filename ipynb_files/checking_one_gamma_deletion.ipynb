```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle

import os
import sys
sys.path.append(os.path.abspath(".."))
from helpers.plotting_3d import generate_box_edge_points
```

```python
with open("../intermediate_files/test_no_deletion.pkl", "rb") as f:
    outputs = pickle.load(f)

no_del_true_gamma_info_df = outputs[0]
no_del_real_nu_reco_nu_downsampled_spacepoints = outputs[1]
no_del_real_nu_reco_cosmic_downsampled_spacepoints = outputs[2]
no_del_real_cosmic_reco_nu_downsampled_spacepoints = outputs[3]
no_del_real_cosmic_reco_cosmic_downsampled_spacepoints = outputs[4]
no_del_real_gamma1_downsampled_spacepoints = outputs[5]
no_del_real_gamma2_downsampled_spacepoints = outputs[6]
no_del_real_other_particles_downsampled_spacepoints = outputs[7]
no_del_real_cosmic_downsampled_spacepoints = outputs[8]

no_del_true_gamma_info_df
```

```python
with open("../intermediate_files/test_all_deletion.pkl", "rb") as f:
    outputs = pickle.load(f)

all_del_true_gamma_info_df = outputs[0]
all_del_real_nu_reco_nu_downsampled_spacepoints = outputs[1]
all_del_real_nu_reco_cosmic_downsampled_spacepoints = outputs[2]
all_del_real_cosmic_reco_nu_downsampled_spacepoints = outputs[3]
all_del_real_cosmic_reco_cosmic_downsampled_spacepoints = outputs[4]
all_del_real_gamma1_downsampled_spacepoints = outputs[5]
all_del_real_gamma2_downsampled_spacepoints = outputs[6]
all_del_real_other_particles_downsampled_spacepoints = outputs[7]
all_del_real_cosmic_downsampled_spacepoints = outputs[8]

all_del_true_gamma_info_df
```

```python
bins = np.linspace(0, 5, 6)
lw=2

plt.figure(figsize=(10, 7))
plt.hist(no_del_true_gamma_info_df["true_num_gamma_pairconvert"], bins=bins, histtype="step", lw=lw, label="No photons deleted")
plt.hist(all_del_true_gamma_info_df["true_num_gamma_pairconvert"], bins=bins, histtype="step", lw=lw, label="One pi0/primary photon deleted")
plt.legend()
plt.xlabel("Number of gamma pair conversions")
plt.ylabel("Number of events")
plt.show()

plt.figure(figsize=(10, 7))
plt.hist(no_del_true_gamma_info_df["true_num_gamma_pairconvert_in_FV"], bins=bins, histtype="step", lw=lw, label="No photons deleted")
plt.hist(all_del_true_gamma_info_df["true_num_gamma_pairconvert_in_FV"], bins=bins, histtype="step", lw=lw, label="One pi0/primary photon deleted")
plt.legend()
plt.xlabel("Number of gamma pair conversions in FV")
plt.ylabel("Number of events")
plt.show()

plt.figure(figsize=(10, 7))
plt.hist(no_del_true_gamma_info_df["true_num_gamma_pairconvert_in_FV_20_MeV"], bins=bins, histtype="step", lw=lw, label="No photons deleted")
plt.hist(all_del_true_gamma_info_df["true_num_gamma_pairconvert_in_FV_20_MeV"], bins=bins, histtype="step", lw=lw, label="One pi0/primary photon deleted")
plt.legend()
plt.xlabel("Number of gamma pair conversions in FV (20 MeV threshold)")
plt.ylabel("Number of events")
plt.show()
```

```python
no_del_reco_nu_vtx = []
all_del_reco_nu_vtx = []
no_del_true_nu_vtx = []
all_del_true_nu_vtx = []
no_del_true_gamma1_pairconvert_vtx = []
no_del_true_gamma2_pairconvert_vtx = []
all_del_true_gamma1_pairconvert_vtx = []
all_del_true_gamma2_pairconvert_vtx = []

for event_i in range(len(no_del_true_gamma_info_df)):

    no_del_reco_nu_vtx.append([no_del_true_gamma_info_df["reco_nu_vtx_x"][event_i], no_del_true_gamma_info_df["reco_nu_vtx_y"][event_i], no_del_true_gamma_info_df["reco_nu_vtx_z"][event_i]])
    no_del_true_nu_vtx.append([no_del_true_gamma_info_df["true_nu_vtx_x"][event_i], no_del_true_gamma_info_df["true_nu_vtx_y"][event_i], no_del_true_gamma_info_df["true_nu_vtx_z"][event_i]])
    all_del_reco_nu_vtx.append([all_del_true_gamma_info_df["reco_nu_vtx_x"][event_i], all_del_true_gamma_info_df["reco_nu_vtx_y"][event_i], all_del_true_gamma_info_df["reco_nu_vtx_z"][event_i]])
    all_del_true_nu_vtx.append([all_del_true_gamma_info_df["true_nu_vtx_x"][event_i], all_del_true_gamma_info_df["true_nu_vtx_y"][event_i], all_del_true_gamma_info_df["true_nu_vtx_z"][event_i]])

    if len(no_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i]) > 0:
        no_del_true_gamma1_pairconvert_vtx.append([no_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i][0], no_del_true_gamma_info_df["true_gamma_pairconversion_ys"][event_i][0], no_del_true_gamma_info_df["true_gamma_pairconversion_zs"][event_i][0]])
    else:
        no_del_true_gamma1_pairconvert_vtx.append(None)
    if len(no_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i]) > 1:
        no_del_true_gamma2_pairconvert_vtx.append([no_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i][1], no_del_true_gamma_info_df["true_gamma_pairconversion_ys"][event_i][1], no_del_true_gamma_info_df["true_gamma_pairconversion_zs"][event_i][1]])
    else:
        no_del_true_gamma2_pairconvert_vtx.append(None)
    if len(all_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i]) > 0:
        all_del_true_gamma1_pairconvert_vtx.append([all_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i][0], all_del_true_gamma_info_df["true_gamma_pairconversion_ys"][event_i][0], all_del_true_gamma_info_df["true_gamma_pairconversion_zs"][event_i][0]])
    else:
        all_del_true_gamma1_pairconvert_vtx.append(None)
    if len(all_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i]) > 1:
        all_del_true_gamma2_pairconvert_vtx.append([all_del_true_gamma_info_df["true_gamma_pairconversion_xs"][event_i][1], all_del_true_gamma_info_df["true_gamma_pairconversion_ys"][event_i][1], all_del_true_gamma_info_df["true_gamma_pairconversion_zs"][event_i][1]])
    else:
        all_del_true_gamma2_pairconvert_vtx.append(None)

if len(no_del_true_gamma1_pairconvert_vtx) == 0:
    no_del_true_gamma1_pairconvert_vtx = None
if len(no_del_true_gamma2_pairconvert_vtx) == 0:
    no_del_true_gamma2_pairconvert_vtx = None
if len(all_del_true_gamma1_pairconvert_vtx) == 0:
    all_del_true_gamma1_pairconvert_vtx = None
if len(all_del_true_gamma2_pairconvert_vtx) == 0:
    all_del_true_gamma2_pairconvert_vtx = None
```

```python
from helpers.plotting_3d import plot_event

event_index = 0

plot_event(event_index, real_gamma1_downsampled_spacepoints=no_del_real_gamma1_downsampled_spacepoints,
                        real_gamma2_downsampled_spacepoints=no_del_real_gamma2_downsampled_spacepoints,
                        real_other_particles_downsampled_spacepoints=no_del_real_other_particles_downsampled_spacepoints,
                        real_cosmic_downsampled_spacepoints=no_del_real_cosmic_downsampled_spacepoints,
                        reco_nu_vtx=no_del_reco_nu_vtx, true_nu_vtx=no_del_true_nu_vtx,
                        true_gamma1_pairconvert_vtx=no_del_true_gamma1_pairconvert_vtx,
                        true_gamma2_pairconvert_vtx=no_del_true_gamma2_pairconvert_vtx)

plot_event(event_index, real_gamma1_downsampled_spacepoints=all_del_real_gamma1_downsampled_spacepoints,
                        real_gamma2_downsampled_spacepoints=all_del_real_gamma2_downsampled_spacepoints,
                        real_other_particles_downsampled_spacepoints=all_del_real_other_particles_downsampled_spacepoints,
                        real_cosmic_downsampled_spacepoints=all_del_real_cosmic_downsampled_spacepoints,
                        reco_nu_vtx=all_del_reco_nu_vtx, true_nu_vtx=all_del_true_nu_vtx,
                        true_gamma1_pairconvert_vtx=all_del_true_gamma1_pairconvert_vtx,
                        true_gamma2_pairconvert_vtx=all_del_true_gamma2_pairconvert_vtx)
```
