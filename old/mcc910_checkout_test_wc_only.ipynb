```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import uproot
```

```python
#f = uproot.open("input_files/checkout_bnb_overlay_test_85files_noweights.root")
f = uproot.open("input_files/wc_mcc910_test_files/checkout_MCC9.10_Test_NuMI_intrinsic_nu_overlay_wirecell_march.root")
f.items()
```

```python
f["wcpselection"].items()
```

```python
bdt_df = f["wcpselection"]["T_BDTvars"].arrays("numu_score", library="pd")

spacepoint_vars = [
    "Trec_spacepoints_x", 
    "Trec_spacepoints_y", 
    "Trec_spacepoints_z", 
    "Trec_spacepoints_q", 
    "Treccharge_spacepoints_x", 
    "Treccharge_spacepoints_y", 
    "Treccharge_spacepoints_z", 
    "Treccharge_spacepoints_q", 
    "Trecchargeblob_spacepoints_x", 
    "Trecchargeblob_spacepoints_y", 
    "Trecchargeblob_spacepoints_z", 
    "Trecchargeblob_spacepoints_q",
]

spacepoints_df = f["wcpselection"]["T_spacepoints"].arrays(spacepoint_vars, library="pd")

total_df = pd.concat([bdt_df, spacepoints_df], axis=1)

total_df
```

```python
numucc_sel_df = total_df.query("numu_score > 0.9")
numucc_sel_df
```

```python
num_points = []
x_points = numucc_sel_df["Trec_spacepoints_x"].array
for i in range(len(numucc_sel_df)):
    num_points.append(len(x_points[i]))


plt.hist(num_points, bins=100, range=(0, 10000))
plt.show()
```

```python
numucc_sel_df["Trec_spacepoints_x"].array[0]
```

```python
event_index = 5

T_rec_x_points = numucc_sel_df["Trec_spacepoints_x"].array[event_index]
T_rec_y_points = numucc_sel_df["Trec_spacepoints_y"].array[event_index]
T_rec_z_points = numucc_sel_df["Trec_spacepoints_z"].array[event_index]
T_rec_q_points = numucc_sel_df["Trec_spacepoints_q"].array[event_index]

T_reccharge_x_points = numucc_sel_df["Treccharge_spacepoints_x"].array[event_index]
T_reccharge_y_points = numucc_sel_df["Treccharge_spacepoints_y"].array[event_index]
T_reccharge_z_points = numucc_sel_df["Treccharge_spacepoints_z"].array[event_index]
T_reccharge_q_points = numucc_sel_df["Treccharge_spacepoints_q"].array[event_index]

T_recchargeblob_x_points = numucc_sel_df["Trecchargeblob_spacepoints_x"].array[event_index]
T_recchargeblob_y_points = numucc_sel_df["Trecchargeblob_spacepoints_y"].array[event_index]
T_recchargeblob_z_points = numucc_sel_df["Trecchargeblob_spacepoints_z"].array[event_index]
T_recchargeblob_q_points = numucc_sel_df["Trecchargeblob_spacepoints_q"].array[event_index]

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(T_rec_y_points, T_rec_x_points, c=T_rec_q_points, s=1)
plt.colorbar(label="Track vs Shower")
plt.xlabel("y (cm)")
plt.ylabel("x (cm)")
plt.title("T_rec spacepoints - XY view")
plt.subplot(122)
plt.scatter(T_rec_z_points, T_rec_x_points, c=T_rec_q_points, s=1)
plt.colorbar(label="Track vs Shower")
plt.xlabel("z (cm)")
plt.ylabel("x (cm)")
plt.title("T_rec spacepoints - ZY view")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(T_reccharge_y_points, T_reccharge_x_points, c=T_reccharge_q_points, s=1)
plt.colorbar(label="Charge")
plt.xlabel("y (cm)")
plt.ylabel("x (cm)")
plt.title("T_reccharge spacepoints - XY view")
plt.subplot(122)
plt.scatter(T_reccharge_z_points, T_reccharge_x_points, c=T_reccharge_q_points, s=1)
plt.colorbar(label="Charge")
plt.xlabel("z (cm)")
plt.ylabel("x (cm)")
plt.title("T_reccharge spacepoints - ZY view")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(T_recchargeblob_y_points, T_recchargeblob_x_points, c=T_recchargeblob_q_points, s=1)
plt.colorbar(label="Charge")
plt.xlabel("y (cm)")
plt.ylabel("x (cm)")
plt.title("T_recchargeblob spacepoints - XY view")
plt.subplot(122)
plt.scatter(T_recchargeblob_z_points, T_recchargeblob_x_points, c=T_recchargeblob_q_points, s=1)
plt.colorbar(label="Charge")
plt.xlabel("z (cm)")
plt.ylabel("x (cm)")
plt.title("T_recchargeblob spacepoints - ZY view")
plt.tight_layout()
plt.show()
```
