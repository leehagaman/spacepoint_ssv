```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import uproot
```

```python
#f = uproot.open("input_files/checkout_bnb_overlay_test_85files_noweights.root")
f = uproot.open("input_files/super_unified_mcc910_test_files/reco_stage_2_hist_c03097ec-924a-45cf-a07d-114ed432b1fc.root")
f.items()
```

```python
categories = [
    "shw_sp",
    "cosmic",
    "gap",
    "mip",
    "pio",
    "sig",
    "mgo",
    "mgt",
    "stw",
    "spt",
    "stem",
    "lem",
    "brm",
    "cme",
    "anc",
    "vis",
    "br",
    "tro",
    "hol",
    "lol",
    "cosmict",
    "numu_cc",
]
cat_counts = {cat: 0 for cat in categories}

for vartup in f["wcpselection"]["T_BDTvars"].items():
    varname = vartup[0]

    if varname in ["run", "subrun", "event", "nuvtx_diff", "showervtx_diff", "muonvtx_diff"]:
        continue
    
    for category in categories:
        if varname.startswith(category):
            cat_counts[category] += 1

for k, v in cat_counts.items():
    print(k, v)
```

```python
numu_scores = f["wcpselection"]["T_BDTvars"].arrays("numu_score", library="np")["numu_score"]

print(numu_scores, np.mean(numu_scores), np.min(numu_scores), np.max(numu_scores))

plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 8))
plt.hist(numu_scores, bins=100, range=(-15, 15))
plt.xlabel("Numu BDT Score")
plt.ylabel("Count")
plt.title("WC Numu BDT Score")
plt.show()
```
