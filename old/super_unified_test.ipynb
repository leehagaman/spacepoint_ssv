```python
pandora_df = f["nuselection"]["NeutrinoSelectionFilter"].arrays(["shr_start_x", "shr_start_y", "shr_start_z", "shr_px", "shr_py", "shr_pz", "blip_x", "blip_y", "blip_z"], library="pd")

pandora_df.head()
```

```python
wc_T_PFeval_df = f["wcpselection"]["T_PFeval"].arrays(["reco_nuvtxX", "reco_nuvtxY", "reco_nuvtxZ", "reco_id"], library="pd")
wc_T_BDTvars_df = f["wcpselection"]["T_BDTvars"].arrays("numu_score", library="pd")

wc_df = pd.concat([wc_T_PFeval_df, wc_T_BDTvars_df], axis=1)
wc_df.head()
```

```python
spacepoint_vars = [
    "Trec_spacepoints_x", 
    "Trec_spacepoints_y", 
    "Trec_spacepoints_z", 
    "Trec_spacepoints_q", 
    "Trec_spacepoints_real_cluster_id",
    "Treccharge_spacepoints_x", 
    "Treccharge_spacepoints_y", 
    "Treccharge_spacepoints_z", 
    "Treccharge_spacepoints_q", 
    "Trecchargeblob_spacepoints_x", 
    "Trecchargeblob_spacepoints_y", 
    "Trecchargeblob_spacepoints_z", 
    "Trecchargeblob_spacepoints_q",
]

spacepoint_df = f["wcpselection"]["T_spacepoints"].arrays(spacepoint_vars, library="pd")

spacepoint_df.head()
```

```python
T_rec_spacepoints_x = spacepoint_df["Trec_spacepoints_x"][4]
T_rec_spacepoints_y = spacepoint_df["Trec_spacepoints_y"][4]
T_rec_spacepoints_z = spacepoint_df["Trec_spacepoints_z"][4]
T_rec_spacepoints_q = spacepoint_df["Trec_spacepoints_q"][4]
T_rec_spacepoints_real_cluster_id = spacepoint_df["Trec_spacepoints_real_cluster_id"][4]

reco_id = wc_df["reco_id"][4]

bins = np.linspace(30000, 50000, 100)

plt.hist(T_rec_spacepoints_real_cluster_id, bins=bins)
plt.hist(reco_id, bins=bins)
plt.yscale("log")

plt.figure()
plt.hist(T_rec_spacepoints_q, bins=100)
plt.yscale("log")

print(T_rec_spacepoints_q)

```

```python
super_unified_df = pd.concat([pandora_df, wc_df, spacepoint_df], axis=1)
super_unified_df.head()
```

```python
overlap_df = super_unified_df.query("shr_start_z > 0")
overlap_df
```

```python
overlap_df[["shr_start_x", "reco_nuvtxX", "shr_start_y", "reco_nuvtxY", "shr_start_z", "reco_nuvtxZ"]]
```

```python
event_index = 2

T_rec_x_points = overlap_df["Trec_spacepoints_x"].array[event_index]
T_rec_y_points = overlap_df["Trec_spacepoints_y"].array[event_index]
T_rec_z_points = overlap_df["Trec_spacepoints_z"].array[event_index]
T_rec_q_points = overlap_df["Trec_spacepoints_q"].array[event_index]

T_reccharge_x_points = overlap_df["Treccharge_spacepoints_x"].array[event_index]
T_reccharge_y_points = overlap_df["Treccharge_spacepoints_y"].array[event_index]
T_reccharge_z_points = overlap_df["Treccharge_spacepoints_z"].array[event_index]
T_reccharge_q_points = overlap_df["Treccharge_spacepoints_q"].array[event_index]

T_recchargeblob_x_points = overlap_df["Trecchargeblob_spacepoints_x"].array[event_index]
T_recchargeblob_y_points = overlap_df["Trecchargeblob_spacepoints_y"].array[event_index]
T_recchargeblob_z_points = overlap_df["Trecchargeblob_spacepoints_z"].array[event_index]
T_recchargeblob_q_points = overlap_df["Trecchargeblob_spacepoints_q"].array[event_index]

blip_x_points = overlap_df["blip_x"].array[event_index]
blip_y_points = overlap_df["blip_y"].array[event_index]
blip_z_points = overlap_df["blip_z"].array[event_index]

pandora_shower_start_x = overlap_df["shr_start_x"].array[event_index]
pandora_shower_start_y = overlap_df["shr_start_y"].array[event_index]
pandora_shower_start_z = overlap_df["shr_start_z"].array[event_index]
pandora_shower_px = overlap_df["shr_px"].array[event_index]
pandora_shower_py = overlap_df["shr_py"].array[event_index]
pandora_shower_pz = overlap_df["shr_pz"].array[event_index]
pandora_direction = np.array([pandora_shower_px, pandora_shower_py, pandora_shower_pz])
pandora_direction = pandora_direction / np.linalg.norm(pandora_direction)

pandora_shower_arrow_length_scaling = 100
pandora_shower_end_x = pandora_shower_start_x + pandora_shower_arrow_length_scaling * pandora_direction[0]
pandora_shower_end_y = pandora_shower_start_y + pandora_shower_arrow_length_scaling * pandora_direction[1]
pandora_shower_end_z = pandora_shower_start_z + pandora_shower_arrow_length_scaling * pandora_direction[2]

print("pandora shower start xyz: ", pandora_shower_start_x, pandora_shower_start_y, pandora_shower_start_z)
print("pandora shower end xyz: ", pandora_shower_end_x, pandora_shower_end_y, pandora_shower_end_z)


plt.figure(figsize=(12, 5))

ax = plt.subplot(121)
plt.scatter(T_recchargeblob_y_points, T_recchargeblob_x_points, c=T_recchargeblob_q_points, s=1, label="WC T_recchargeblob spacepoints")
plt.scatter(blip_y_points, blip_x_points, s=100, marker="*", c="red", zorder=-1, label="blips")

ax.annotate(
    "", 
    xy=(pandora_shower_end_y, pandora_shower_end_x),      # Arrowhead (ending point)
    xytext=(pandora_shower_start_y, pandora_shower_start_x),  # Arrow tail (starting point)
    arrowprops=dict(arrowstyle="->", color='tab:orange', linewidth=2,),  # Arrow style properties
    zorder=100
)

plt.colorbar(label="Charge")
plt.xlabel("y (cm)")
plt.ylabel("x (cm)")
plt.title("T_recchargeblob spacepoints - XY view")

ax = plt.subplot(122)
plt.scatter(T_recchargeblob_z_points, T_recchargeblob_x_points, c=T_recchargeblob_q_points, s=1, label="WC T_recchargeblob spacepoints")
plt.scatter(blip_z_points, blip_x_points, s=100, marker="*", c="red", zorder=-1, label="blips")
ax.annotate(
    "", 
    xy=(pandora_shower_end_z, pandora_shower_end_x),      # Arrowhead (ending point)
    xytext=(pandora_shower_start_z, pandora_shower_start_x),  # Arrow tail (starting point)
    arrowprops=dict(arrowstyle="->", color='tab:orange', linewidth=2,),  # Arrow style properties
    zorder=-1,
    label="pandora shower",
)
plt.plot([], [], color='tab:orange', linewidth=3, label="Pandora Reconstructed Shower")
plt.colorbar(label="Charge")
plt.legend(loc="upper right")
plt.xlabel("z (cm)")
plt.ylabel("x (cm)")
plt.title("T_recchargeblob spacepoints - ZY view")
plt.tight_layout()
plt.show()
```
